```python
import oracledb
from logowanie_baza import user, password, dsn

conn = oracledb.connect(user=user,
                        password=password,
                        dsn=dsn)

print('Connected')
sql = 'select * from tasks_mg'
c = conn.cursor()
c.execute(sql)
for id, name, descr, created, due in c:
    print(name, due)
c.close()
conn.close()
```
