# Praca z bazami Oracle
Praca z bazą Oracle odbywa się z użyciem jednej z 2 bibliotek:
- cx_Oracle - stara biblioteka używana w sqlalchemy w wersjach przed 2.0
- oracledb - nowa biblioteka

## Biblioteka cx_Oracle

Biblioteka cx_oracle wymaga instalacji Oracle Instant Client we w miarę aktualnej wersji.
Bibliotekę można pobrać tu:

https://www.oracle.com/database/technologies/instant-client/downloads.html

Po pobraniu należy ją:
1. Rozpakować i skopiować pełną ścieżkę do katalogu, np `C:\dev\instantclient`
2. Dodać ścieżkę do zmiennej PATH.
3. Zweryfikować poprrawność uruchamiając konsolę i wpisując polecenie 'genezi'. Jeżeli uruchomi się i wypisze podpowiedź - to znaczy że poprawnie ustawiliśmy PATH.
4. Zrestartować Jupytera wrac z Anacondą.
5. Uruchomić ponownie.
6. Uruchomić poniższy kod. On się nie połączy, ale jeżeli zobaczymy komunikat że instant client not found, to znaczy że coś zrobiliśmy źle w powyższych punktach.



```python
import cx_Oracle
user = 'blabla'
password = 'blablapasswd'
dsn = 'localhost:2233/niematakiejbazy'
try:
    conn = cx_Oracle.connect(user=user,
                         password=password,
                         dsn=dsn)
except cx_Oracle.DatabaseError as e:
    print('Błąd połączenia jest ok, ale jeżeli nie widzi Instant Client to wracamy do powyższych punktów.')
    print('Błąd:')
    print(type(e).__name__, ':', e)
```

Następnie uruchamiamy notatniki demostrujące pracę z bazą:
- bazy_01_cx_oracle - tworzenie prostej bazy na poziomie dbapi 2.0
- bazy_02_cx_oracle - minimalny przykład odczytu danych z bazy przy użyciu dbapi 2.0
- bazy_03_cx_oracle - przykład demonstrujący Object Relational Mapping (ORM) w paczce sqlalchemy
- bazy_04_cx_oracle - przykład demonstrujący pracę z poziomu pandas

W każym przypadku musimy znać swojego usera i hasło.

zazwyczaj nadawane są one wg schematu:
- `Sluchacz1` do `SluchaczN`
- `Sluchacz1_passwd` do `SluchaczN_passwd`

gdzie N to ilość słuchaczy.

Dsn do bazy i nazwy użytkowników są ustawione w pliku logowanie_baza.py. Należy sobie otworzyć ten plik, wybrać wolny nr usera i ustawić jego nazwę i hasło.


A teraz spróbujemy połączyć się z realną bazą. Popraw w pliku logowanie_baza.py dane logowanie i próbuj poniższego kodu.

```python
import cx_Oracle
from logowanie_baza import user, password, dsn

with cx_Oracle.connect(user=user,
                         password=password,
                         dsn=dsn) as conn:
    print(conn.username, conn.dsn)
```

## Biblioteka oracledb

Biblioteka oracledb zastąpiła starą bibliotekę cx_Oracle.

Nie wymaga ona instalacji Oracle Instant Client jeżeli działa w domyśnym trybie Thin Client. Jeżeli chcemy jednak używać Oracle Instant Client to należy ją uruchamiać w trybie Thick Client.

W wczesnych wersjach, biblioteka wymagała pliku tsnnames.ora zawierającego dane serwerów bazodanowych. Opis użycia i konfiguracji:
https://python-oracledb.readthedocs.io/en/latest/user_guide/connection_handling.html#netservice

W nowszych wersjach nie jest to konieczne i parametry połączenia możemy podać ręcznie.
Spróbujmy zatem połączyć się z bazą:

```python
import oracledb
from logowanie_baza import user, password, dsn

with oracledb.connect(user=user,
                         password=password,
                         dsn=dsn) as conn:
    print(conn.username, conn.dsn)
```
