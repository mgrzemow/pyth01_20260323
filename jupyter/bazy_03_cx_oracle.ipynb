```python
import random
from pprint import pprint
from logowanie_baza import user, password, dsn

import faker as fk
import sqlalchemy as db
from sqlalchemy import create_engine
from sqlalchemy.orm import declarative_base
from sqlalchemy.orm import sessionmaker
dsn_1 = 'oracle+cx_oracle://{}:{}@{}'

# do wczytywania haseł:
import getpass
magiczny_str_polaczeniowy = dsn_1.format(user, password, dsn)
print(magiczny_str_polaczeniowy)
engine = create_engine(magiczny_str_polaczeniowy)

Session = sessionmaker(engine)
session = Session()
Base = declarative_base()


class Czlowiek(Base):
    id = 1
    __tablename__ = 'czlowiek_mg'
    czlowiekId = db.Column(db.Integer, name='czlowiek_id', primary_key=True)
    imie = db.Column(db.String(100), name='imie')
    nazwisko = db.Column(db.String(100), name='nazwisko')
    wiek = db.Column(db.Integer, name='wiek')
    wzrost = db.Column(db.Float, name='wzrost')

    def __init__(self, imie, nazwisko, wiek, wzrost):
        self.czlowiekId = Czlowiek.id
        Czlowiek.id += 1
        self.imie = imie
        self.nazwisko = nazwisko
        self.wiek = wiek
        self.wzrost = wzrost

    def __str__(self):
        return f'Czlowiek({self.czlowiekId}, {self.imie}, {self.nazwisko}, {self.wiek} lat, {self.wzrost} cm.)'

    def __repr__(self):
        return str(self)



def generate_data(nr=10):
    f = fk.Faker('pl_PL')
    lista = []
    for _ in range(nr):
        lista.append(Czlowiek(f.first_name(),
                              f.last_name(),
                              random.randint(25, 89),
                              random.randint(1500, 1900) / 10))
    # tu mam obiekty wyłącznie lokalne
    print('New objects before session.add:')
    pprint(session.new)

    for c in lista:
        session.add(c)
    print('New objects  after session.add:')
    pprint(session.new)
    session.commit()
    print('New objects after session.commit:')
    pprint(session.new)


def init_db():
    Base.metadata.drop_all(engine)
    Base.metadata.create_all(engine)


if __name__ == '__main__':
    init_db()
    generate_data(30)
    lista_osob = session.query(Czlowiek).filter(Czlowiek.wzrost >= 170).all()
    for o in lista_osob:
        o.wzrost = round(o.wzrost)
    # Dirty - obiekty ze zmienionymi wartościami jeszcze nie zapisane w bazie
    print('Dirty: ', session.dirty)
    session.commit()
    print('Dirty: ', session.dirty)
    lista_osob = session.query(Czlowiek).filter(Czlowiek.wzrost >= 170).all()
    print(lista_osob)

    print('Usuwam powyżej 80 lat:')
    print(session.query(Czlowiek).filter(Czlowiek.wiek >= 80))
    print(session.query(Czlowiek).filter(Czlowiek.wiek >= 80).all())
    lista_osob = session.query(Czlowiek).filter(Czlowiek.wiek >= 80).delete()
    session.commit()
    print('Po usunięciu: ')
    print(session.query(Czlowiek).filter(Czlowiek.wiek >= 80).all())

```
