```python
import datetime as dt
import cx_Oracle
import faker as fk
from logowanie_baza import user, password, dsn



def stworz(conn):
    kursor = conn.cursor()
    sql_create = """CREATE TABLE tasks_mg (
    	task_id NUMBER(10) PRIMARY KEY,
    	name VARCHAR2(50) UNIQUE NOT NULL,
    	description VARCHAR2(1000) UNIQUE NOT NULL,
    	created_date TIMESTAMP NOT NULL,
        due_date TIMESTAMP)
        """

    kursor.execute(sql_create)
    conn.commit()
    kursor.close()


def zniszcz(conn):
    kursor = conn.cursor()
    sql_create = """  BEGIN
      EXECUTE IMMEDIATE 'DROP TABLE tasks_mg';
  EXCEPTION
      WHEN OTHERS THEN NULL;
  END;"""
    kursor.execute(sql_create)
    conn.commit()
    kursor.close()

def insert_data(kursor, id, name, description, created_d, due_d):
    sql_insert = f"""insert into tasks_mg (task_id, name, description, created_date, due_date) 
                                   values (:task_id, :name, :description, :created_date, :due_date)"""
    kursor.execute(sql_insert, (id, name, description, created_d, due_d))
    return id


def fill_data(conn, nr=100):
    f = fk.Faker()
    f = fk.Faker('pl_PL')
    f.add_provider(fk.providers.company.Provider)
    with conn.cursor() as c:
        for i in range(nr):
            insert_data(c, i, f.bs(), f.catch_phrase(), dt.datetime.now(), dt.datetime.now() + dt.timedelta(days=7))
        conn.commit()


if __name__ == '__main__':

    with cx_Oracle.connect(user=user,
                             password=password,
                             dsn=dsn) as conn:
        zniszcz(conn)
        stworz(conn)
        fill_data(conn, 100)

        sql = 'select * from tasks_mg'
        with conn.cursor() as c:
            c.execute(sql)
            for id, name, descr, created, due in c:
                print(name, due)
```
