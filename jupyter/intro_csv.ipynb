```python
import pandas as pd
```

```python
df = pd.read_csv('requests.csv', parse_dates=['Created Date'], low_memory=False)
```

```python
df.head().to_dict()
```

```python
parties = df[df['Descriptor'] == 'Loud Music/Party']
```

```python
parties['Created Date'].dt.weekday.value_counts().sort_index().plot(kind='bar')
```

```python
insects = df[df['Descriptor'] == 'VERMIN']
```

```python
insects['Borough'].value_counts().plot(kind='bar')
```

```python
boroughs_population = pd.Series({
    'BRONX': 1385107,
    'BROOKLYN': 2504706,
    'MANHATTAN': 1585874,
    'QUEENS': 2230545,
    'STATEN ISLAND': 468730,
})
```

```python
insects_per_person = insects['Borough'].value_counts() / boroughs_population
```

```python
insects_per_person.sort_values(ascending=False).plot(kind='bar')
```
